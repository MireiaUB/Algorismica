# Problemes de funcions de cadenes, islower, isupper, isalpha, lower i upper

**1**. Comprova si un caràcter és lletra minúscula o majúscula

```python
#@title Solució
def comprovar_lletra(caracter):
    if caracter.islower():
        print("És una lletra minúscula.")
    elif caracter.isupper():
        print("És una lletra majúscula.")
    else:
        print("No és una lletra.")
```

```python
caracter = input("Introdueix un caràcter: ")
comprovar_lletra(caracter)
```

**2**. Comprova si una cadena (string) conté només lletres. Prova d'introduir diversos textos fins que et retorni que la cadena només té lletres.

```python
#@title Solució

def comprovar_nomes_lletres(cadena):
    if cadena.isalpha():
        print("La cadena conté només lletres.")
    else:
        print("La cadena conté caràcters no alfabètics.")
```

```python
text = input("Introdueix un text: ")
comprovar_nomes_lletres(text)
```

**3**. Converteix una cadena a majúscules

```python
#@title Solució
def convertir_a_majuscules(cadena):
    majuscules = cadena.upper()
    print("La cadena en majúscules és:", majuscules)
```

```python
text = input("Introdueix un text: ")
convertir_a_majuscules(text)
```

**4**. Converteix una cadena a minúscules

```python
#@title Solució

def convertir_a_minuscules(cadena):
    minuscules = cadena.lower()
    print("La cadena en minúscules és:", minuscules)
```

```python
text = input("Introdueix un text: ")
convertir_a_minuscules(text)
```

**5**. Comprova si una cadena conté lletres minúscules

```python
#@title Solució

def comprovar_minuscules(cadena):
    if any(caracter.islower() for caracter in cadena):
        print("La cadena conté lletres minúscules.")
    else:
        print("La cadena no conté lletres minúscules.")
```

```python
text = input("Introdueix un text: ")
comprovar_minuscules(text)
```

**11**. Comprova si una cadena conté només lletres i espais

```python
#@title Solució

def comprovar_lletres_i_espais(cadena):
    if all(caracter.isalpha() or caracter.isspace() for caracter in cadena):
        print("La cadena conté només lletres i espais.")
    else:
        print("La cadena conté caràcters no alfabètics o espais.")
```

```python
text = input("Introdueix un text: ")
comprovar_lletres_i_espais(text)
```

**12**. Comprovar si una cadena conté només lletres i dígits.

```python
#@title Solució
def comprovar_lletres_i_digits(cadena):
    if all(caracter.isalpha() or caracter.isdigit() for caracter in cadena):
        print("La cadena conté només lletres i dígits.")
    else:
        print("La cadena conté caràcters no alfabètics o dígits.")
```

```python
text = input("Introdueix un text: ")
comprovar_lletres_i_digits(text)
```

**5**. Escriu una funció _area_cercle(radi)_ que calculi l'àrea d'un cercle donat el seu radi.

Si et cal importa la llibreria math i usa alguna de les seves constants.

**6**. Escriu una funció _celsius_a_fahrenheit(celsius)_  que converteixi graus Celsius a Fahrenheit.

```python
#@title Solució
def celsius_a_fahrenheit(celsius):
    return (celsius * 9/5) + 32
```

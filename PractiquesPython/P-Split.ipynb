# Problemes amb split

**1**. Escriu una funció que prengui una cadena i la divideixi en paraules, utilitzant l'espai en blanc com a delimitador.

Exemple

Input: "Hola com estàs?"

Output: ["Hola", "com", "estàs?"]

```python
#@title Solució
def divideix_paraules(cadena):
    return cadena.split()

frase = "Hola com estàs?"
print(divideix_paraules(frase))
```

**2**. Escriu una funció que prengui una cadena amb noms separats per comes i la divideixi en una llista de noms.

Exemple:

Input: "Maria,Anna,Joan"

Output: ["Maria", "Anna", "Joan"]

```python
#@title Solució

def divideix_noms(cadena):
    return cadena.split(',')

noms = "Maria,Anna,Joan"
print(divideix_noms(noms))
```

**3**. Escriu una funció que prengui una cadena de nombres separats per guions i la divideixi en una llista de nombres enters.

Exemple:

Input: "10-20-30-40"
Output: [10, 20, 30, 40]

```python
#@title Solució
def divideix_nombres(cadena):
    nombres = cadena.split('-')
    nombres_enters = [int(x) for x in nombres]
    return nombres_enters

cadena_nombres = "10-20-30-40"
print(divideix_nombres(cadena_nombres))
```

**4**. Escriu una funció que prengui una cadena de paraules separades per guions baixos i la divideixi en una llista de paraules.

Exemple:

Input: "aquest_es_un_text_de_prova"

Output: ["aquest", "es", "un", "text", "de", "prova"]

```python
#@title Solució

def divideix_text(cadena):
    return cadena.split('_')

text = "aquest_es_un_text_de_prova"
print(divideix_text(text))
```

**5**. Escriu una funció que prengui una cadena de dates en format "dia/mes/any" i la divideixi en una llista de tuples de la forma (dia, mes, any).

Exemple:

Input: "20/07/2023-15/12/2022-05/03/2021"

Output: [("20", "07", "2023"), ("15", "12", "2022"), ("05", "03", "2021")]

```python
#@title Solució

def divideix_dates(cadena):
    dates = cadena.split('-')
    dates_dividides = [tuple(date.split('/')) for date in dates]
    return dates_dividides

dates_input = "20/07/2023-15/12/2022-05/03/2021"
print(divideix_dates(dates_input))
```

**6**. Escriu una funció que prengui una cadena de text i la divideixi en una llista de frases separades per punts.

Exemple:

Input: "Hola. Com estàs? Bona nit."

Output: [["Hola"], [" Com estàs? Bona nit."]]


```python
#@title Solució

def divideix_frases_en_paraules(cadena):
    frases = cadena.split('. ')
    llistes_paraules = [frase.split() for frase in frases]
    return llistes_paraules
```

```python
divideix_frases_en_paraules( "Hola. Com estàs? Bona nit.")
```

**7**. Escriu una funció que prengui una cadena de text i la divideixi en una llista de línies.

Exemple:

Input:

"Hola

Com estàs?

Bona nit."

Output: ["Hola", "Com estàs?", "Bona nit."]


```python
#@title Solució

def divideix_linies(cadena):
    return cadena.split('\n')

text = "Hola\nCom estàs?\nBona nit."
print(divideix_linies(text))
```

**8**. Escriu una funció que prengui una cadena amb números decimals separats per guions i la divideixi en una llista de nombres de punt flotant.

Exemple:

Input: "3.14-2.71-1.618"

Output: [3.14, 2.71, 1.618]

```python
#@title Solució

def divideix_floats(cadena):
    numeros = cadena.split('-')
    numeros_floats = [float(x) for x in numeros]
    return numeros_floats

cadena_floats = "3.14-2.71-1.618"
print(divideix_floats(cadena_floats))
```

**9**. Escriu una funció que prengui una cadena amb noms i edats separades per espais i la divideixi en una llista de tuples de la forma (nom, edat).

Exemple:

Input: "Maria 25 Joan 30 Anna 28"

Output: [("Maria", "25"), ("Joan", "30"), ("Anna", "28")]

```python
#@title solució
def divideix_noms_i_edats(cadena):
    elements = cadena.split()
    tuples_nom_edat = [(elements[i], elements[i+1]) for i in range(0, len(elements), 2)]
    return tuples_nom_edat

cadena_nom_edat = "Maria 25 Joan 30 Anna 28"
print(divideix_noms_i_edats(cadena_nom_edat))
```

**1**. Donada la següent cadena, eliina els espais en blanc al principi i al final:


```python
text = "   Hola, què tal?   "
```

```python
#@title Solució
text_net = text.strip()
print(text_net)
```

**2**. Escriu una funció que donada una llista de cadenes elimini els espais en blanc al principi i al final de cada cadena a la llista.

```python
#@title Solució

def netejar_cadenes(llista_cadenes):
    return [cadena.strip() for cadena in llista_cadenes]
```

```python
cadenes = ["   Poma  ", "  Taronja ", "   Plàtan"]
cadenes_net = netejar_cadenes(cadenes)
print(cadenes_net)
```

**3**. Escriu una funció que donada una contrasenya la validi amb una que ja té guardada la funció (contrasenya_bd que és "laMevaContrasenya123"), ignorant espais al principi i al final.

```python
#@title Solució
def verificar_contrasenya(contrasenya):
    contrasenya_bd = " laMevaContrasenya123"
    return contrasenya.strip() == contrasenya_bd.strip()
```

```python
contrasenya = "   laMevaContrasenya123   "
if verificar_contrasenya(contrasenya):
    print("Contrasenya válida.")
else:
    print("Contrasenya incorrecta.")
```

**4**. Escriu una funció que donada una cadena retorni True si comença i acaba amb el mateix caràcter, ignorant els espais en blanc al principi i al final.

```python
#@title Solució

def mateix_caracter(cadena):
    cadena_neta = cadena.strip()
    return cadena_neta[0] == cadena_neta[-1]
```

```python
cadena1 = "  abcba "
cadena2 = " abcda "
print(mateix_caracter(cadena1))
print(mateix_caracter(cadena2))
```

# Problemes amb diferents tipus de variables

**1**. Suma de dos nombres enters

Demana a l'usuari dos nombres enters i mostra la seva suma

```python
#@title Solució
def suma_dos_enters():
    num1 = int(input("Introduïu el primer nombre enter: "))
    num2 = int(input("Introduïu el segon nombre enter: "))
    suma = num1 + num2
    print("La suma és:", suma)
```

```python
suma_dos_enters()
```

**2**. Mitjana de tres nombres decimals

Demana a l'usuari tres nombres decimals i calcula la mitjana.


```python
#@title Solució

def mitjana_tres_decimals():
    num1 = float(input("Introduïu el primer nombre decimal: "))
    num2 = float(input("Introduïu el segon nombre decimal: "))
    num3 = float(input("Introduïu el tercer nombre decimal: "))
    mitjana = (num1 + num2 + num3) / 3
    print("La mitjana és:", mitjana)
```

```python
mitjana_tres_decimals()
```

**3**. Conversió d'hores, minuts i segons a segons

Demana a l'usuari un temps en hores, minuts i segons, i calcula la seva equivalència en segons.

```python
#@title Solució
def temps_a_segons():
    hores = int(input("Introduïu les hores: "))
    minuts = int(input("Introduïu els minuts: "))
    segons = int(input("Introduïu els segons: "))

    total_segons = (hores * 3600) + (minuts * 60) + segons
    print("El temps en segons és:", total_segons)
```

```python
temps_a_segons()
```

**4**. Càlcul de l'edat mitjana d'un grup de persones

Demana a l'usuari les edats de 5 persones i calcula l'edat mitjana.

```python
#@title Solució

def edat_mitjana():
    edats = []
    for i in range(5):
        edat = int(input(f"Introduïu l'edat de la persona {i + 1}: "))
        edats.append(edat)
    mitjana = sum(edats) / len(edats)
    print("L'edat mitjana és:", mitjana)
```

```python
edat_mitjana()
```

**5**. Converteix un nombre decimal a binari

```python
#@title Solució

decimal = int(input("Introdueix un nombre decimal: "))
binari = bin(decimal)
```

```python
print(f"El nombre {decimal} en binari es: {binari}")
# fixa't en com en aquest print es fusionen els valors de les variables amb el text
```

# Problemes de slicing

**1**. Donada la cadena "Hello World", imprimeix les dues primeres lletres.

```python
#@title Solució
cadena = "Hello World"
print(cadena[:2])
```

**2**. Donada la cadena "Python Programming", imprimeix les lletres des de la tercera fins a la sisena.

```python
#@title Solució

cadena = "Python Programming"
print(cadena[2:6])
```

**3**. Donada la cadena "Engineering", imprimeix els últims tres caràcters.

```python
#@title Solució
cadena = "Engineering"
print(cadena[-3:])
```

**4**. Donada la cadena "Algorithms' Subject is awesome", imprimeix la part central (sense les dues primeres i les dues últimes lletres).

```python
#@title Solució

cadena = "Algorithms' Subject is awesome"
print(cadena[2:-2])
```

**5**. Donada la cadena "Python is fun", imprimeix les lletres saltant de dos en dos.

```python
#@title Solució

cadena = "Python is fun"
print(cadena[::2])
```

**6**. Donada la cadena "Hello, World!", imprimeix-la al revés.

```python
#@title Solució
cadena = "Hello, World!"
print(cadena[::-1])
```

**7**. Donada la cadena "123456789", imprimeix només els números parells.

```python
#@title Solució

cadena = "123456789"
parells = cadena[1::2]
print(parells)
```

**8**. Donada la cadena "Python", imprimeix-la en ordre invers i amb guions entre les lletres.

```python
#@title Solució
cadena = "Python"
inversa = '-'.join(cadena[::-1])
print(inversa)
```

**9**. Donada la cadena "Welcome to Python", imprimeix-la centrant-la en una línia de 30 caràcters.

```python
#@title solució

cadena = "Welcome to Python"
centrada = cadena.center(30)
print(centrada)

# Aquesta funció no l'havíem vist, però aixi veieu que Python té possibibilitats infinites.
```

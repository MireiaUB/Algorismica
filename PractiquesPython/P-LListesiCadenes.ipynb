**1**. Escriu una funció que agafi dues llistes d'enters i retorni una nova llista que contingui la suma dels elements corresponents de les dues llistes.

```python
#@title Solució
def suma_llistes(llista1, llista2):
    suma = []
    for i in range(len(llista1)):
        suma.append(llista1[i] + llista2[i])
    return suma
```

```python
# Exemple d'ús
llista1 = [1, 2, 3]
llista2 = [4, 5, 6]
resultat = suma_llistes(llista1, llista2)
print(resultat)  # Sortida: [5, 7, 9]
```

**10**. Escriu una funció que prengui una llista d'enters i retorni dues llistes separades, una amb els nombres parells i l'altra amb els nombres senars.

```python
#@title Solució
def separar_parells_senars(llista):
    parells = [num for num in llista if num % 2 == 0]
    senars = [num for num in llista if num % 2 != 0]
    return parells, senars
```

```python
# Exemple d'ús
llista = [1, 2, 3, 4, 5, 6]
parells, senars = separar_parells_senars(llista)
print(parells)  # Sortida: [2, 4, 6]
print(senars)   # Sortida: [1, 3, 5]
```

**16**. Escriu una funció que sumi tots els elements d'una llista de nombres.

```python
#@title Solució amb funció específica

def sumar_elements(llista):
    suma = sum(llista)
    return suma

# Ho podries fer amb una iteració?
```

```python
#@title Solució amb iteració

def sumar_elements_iteracio(llista):
    suma = 0
    for element in llista:
        suma += element
    return suma
```

```python
# Exemple d'ús
llista = [1, 2, 3, 4, 5]
resultat = sumar_elements(llista)
print(resultat)  # Sortida: 15
llista = [2, 3, 4, 5, 6]
resultat = sumar_elements_iteracio(llista)
print(resultat)  # Sortida: 20
```

**28**. Càlcul d'àrees
Escriu una funció que calculi l'àrea d'un cercle o un rectangle. La funció rebrà el tipus de figura ('cercle' o 'rectangle') i els paràmetres necessaris per al càlcul.

```python
#@title Solució

import math

def calcula_area(figura, *args):
    if figura == 'cercle':
        radi = args[0]
        return math.pi * radi ** 2
    elif figura == 'rectangle':
        base, altura = args
        return base * altura
    else:
        return None
```

```python
# Exemple d'ús:
area_cercle = calcula_area('cercle', 5)
area_rectangle = calcula_area('rectangle', 3, 4)
print(area_cercle)  # Mostrarà 78.53981633974483
print(area_rectangle)  # Mostrarà 12
```

**29**. Comptar vocals

Escriu una funció que compti el nombre de vocals en una cadena de text i retorni el resultat.

```python
#@title Solució

def comptar_vocals(cadena):
    # Inicialitzem un comptador per a les vocals
    comptador = 0

    # Recorrem cada caracter en la cadena
    for caracter in cadena:
        # Convertim el caracter a minúscules
        # per fer la comparació insensible a majúscules o minúscules
        caracter = caracter.lower()

        # Comprovem si el caracter es una vocal
        if caracter in 'aeiouáéíóú':
            comptador += 1

    return comptador
```

```python
# Exemple d'ús
cadena = "Hola, aquesta és una cadena d'exemple"
nombre_de_vocals = comptar_vocals(cadena)
print("El nombre de vocals a la cadena es:", nombre_de_vocals)
```

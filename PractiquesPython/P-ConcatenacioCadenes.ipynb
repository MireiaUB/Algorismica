**2**. Escriu un programa que demani a l'usuari que introdueixi dues paraules i les mostri concatenades.

```python
#@title Solució

paraula1 = input("Introdueix la primera paraula: ")
paraula2 = input("Introdueix la segona paraula: ")
concatenacio = paraula1 + " " + paraula2
print("La concatenació és:", concatenacio)
```

**3**. Escriu un programa que demani a l'usuari que introdueixi una paraula i un número, i mostri la paraula repetida tantes vegades com indiqui el número.

```python
#@title Solució
paraula = input("Introdueix una paraula: ")
numero = int(input("Introdueix un número: "))
repetit = paraula * numero
print("La paraula repetida és:", repetit)
```

**4**. Escriu un programa que demani a l'usuari que introdueixi tres paraules i mostri les tres paraules concatenades sense espais.

```python
#@title Solució

paraula1 = input("Introdueix la primera paraula: ")
paraula2 = input("Introdueix la segona paraula: ")
paraula3 = input("Introdueix la tercera paraula: ")
concatenacio = paraula1 + paraula2 + paraula3
print("La concatenació és:", concatenacio)
```

**5**. Escriu un programa que agafi dues paraules com a entrada i mostri la primera paraula seguida de la segona paraula repetida 5 vegades, separades per un espai.

```python
#@title Solució

paraula1 = input("Introdueix la primera paraula: ")
paraula2 = input("Introdueix la segona paraula: ")
repetit = paraula1 + " " + (paraula2 + " ") * 5
print("La paraula repetida és:", repetit)
```

**6**. Escriu un programa que demani a l'usuari que introdueixi el seu nom i mostri el seu nom repetit 10 vegades en línies diferents.

```python
#@title Solució
nom = input("Introdueix el teu nom: ")
for i in range(10):
    print(nom)

# Si vols explorar el mateix resultat també es podria fer amb una única cadena
# i un caràcter especial per fer el salt de línia. Prova-ho!
```

**7**. Escriu un programa que demani a l'usuari que introdueixi dues paraules i mostri la primera paraula amb lletres majúscules i la segona amb lletres minúscules, tot concatenant-les.

```python
#@title Solució

paraula1 = input("Introdueix la primera paraula: ")
paraula2 = input("Introdueix la segona paraula: ")
concatenacio = paraula1.upper() + paraula2.lower()
print("La concatenació és:", concatenacio)
```

# Problemes if/else

**1**. Escriu un programa que demani a l'usuari un nombre i mostri si és positiu, negatiu o zero.

```python
#@title Solució
num = int(input("Introdueix un nombre: "))

if num > 0:
    print("El nombre és positiu.")
elif num < 0:
    print("El nombre és negatiu.")
else:
    print("El nombre és zero.")
```

**2**. Escriu un programa que demani a l'usuari dues notes (entre 0 i 100) i mostri la seva mitjana.

```python
#@title Solució

nota1 = int(input("Introdueix la primera nota: "))
nota2 = int(input("Introdueix la segona nota: "))

if 0 <= nota1 <= 100 and 0 <= nota2 <= 100:
    mitjana = (nota1 + nota2) / 2
    print("La mitjana és:", mitjana)
else:
    print("Les notes han de ser entre 0 i 100.")
```

**3**. Escriu un programa que comprovi si un nombre és parell o imparell

```python
#@title Solució
num = int(input("Introdueix un nombre: "))

if num % 2 == 0:
    print("El nombre és parell.")
else:
    print("El nombre és imparell.")
```

**4**. Escriu un programa que demani a l'usuari dues paraules i mostri quina paraula és més llarga.

```python
#@title Solució

paraula1 = input("Introdueix la primera paraula: ")
paraula2 = input("Introdueix la segona paraula: ")

if len(paraula1) > len(paraula2):
    print("La primera paraula és més llarga.")
elif len(paraula1) < len(paraula2):
    print("La segona paraula és més llarga.")
else:
    print("Les dues paraules tenen la mateixa longitud.")

# Fixeu-vos en l'ús d'elif
```

**5**. Escriu un programa que demani a l'usuari el seu edat i digui si és menor d'edat (menys de 18 anys) o major d'edat.

```python
#@title Solució

edat = int(input("Quina és la teva edat? "))

if edat < 18:
    print("Ets menor d'edat.")
else:
    print("Ets major d'edat.")
```

**6**. Escriu un programa que determini si un any donat és un any de traspàs.

```python
#@title Solució
any = int(input("Introdueix un any: "))

if any % 4 == 0 and (any % 100 != 0 or any % 400 == 0):
    print("És un any de traspàs.")
else:
    print("No és un any de traspàs.")
```

**7**. Escriu un programa que determini si un triangle és equilàter, isòsceles o escalè.

```python
#@title Solució

costat1 = float(input("Introdueix la longitud del primer costat: "))
costat2 = float(input("Introdueix la longitud del segon costat: "))
costat3 = float(input("Introdueix la longitud del tercer costat: "))

if costat1 == costat2 == costat3:
    print("El triangle és equilàter.")
elif costat1 == costat2 or costat1 == costat3 or costat2 == costat3:
    print("El triangle és isòsceles.")
else:
    print("El triangle és escalè.")
```

**8**. Escriu un programa que demani l'any de naixement de l'usuari i mostri la seva categoria d'edat (p.ex. infantil, juvenil, adult).

**9**. Escriu un programa que demani l'edat de tres persones i mostri la persona més gran i la més petita.

```python
#@title solució
edat1 = int(input("Introdueix l'edat de la primera persona: "))
edat2 = int(input("Introdueix l'edat de la segona persona: "))
edat3 = int(input("Introdueix l'edat de la tercera persona: "))

edats = [edat1, edat2, edat3]
edats.sort()

print("La persona més petita té", edats[0], "anys.")
print("La persona més gran té", edats[-1], "anys.")

# Se t'acudeix una solució sense utilitzar sort?
```

**10**. Escriu un programa que converteixi una nota numèrica en una qualificació amb lletres, seguint la següent taula:

+ 9-10: Excel·lent
+ 7-8: Notable
+ 5-6: Aprovat
+ 0-4: Suspes

```python
#@title Solució
nota = float(input("Introdueix la teva nota: "))

if 9 <= nota <= 10:
    qualificacio = "Excel·lent"
elif 7 <= nota < 9:
    qualificacio = "Notable"
elif 5 <= nota < 7:
    qualificacio = "Aprovat"
print("Qualificació: ", qualificacio)
```
